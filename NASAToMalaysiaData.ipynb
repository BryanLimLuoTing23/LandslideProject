```python

import pandas as pd

# Load the NASA landslide dataset
nasa_landslide_data = pd.read_csv('nasa_global_landslide_catalog_point.csv')


nasa_landslide_data['event_date'] = pd.to_datetime(nasa_landslide_data['event_date'], errors='coerce')

#Remove all rows with missing values in these essential columns
essential_columns = ['latitude', 'longitude', 'event_date', 'landslide_size']
cleaned_data = nasa_landslide_data.dropna(subset=essential_columns)
```

###Test for Malaysia data for now, I'll assume the location accuracy is perfect. This is a possible improvement for later.

```python
columns_to_drop = [
    'event_id', 'landslide_setting', 'injury_count', 'storm_name', 'photo_link', 'event_import_source','fatality_count', 'comments', 'country_code','landslide_category',
    'OBJECTID', 'source_name', 'event_title', 'event_description', 
    'location_description', 'Shape', 'source_link', 
    'event_time', 'last_edited_date'
]

cleaned_data = cleaned_data.drop(columns = columns_to_drop, errors = 'ignore')

cleaned_data.head()
```

Extract Malaysian data

```python
malaysia_data = cleaned_data[cleaned_data['country_name'] == 'Malaysia']

# Display the first few rows of the filtered dataset
malaysia_data
```

```python
malaysia_data.to_csv('Malaysia_Data.csv', index=False)

```
